```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Crop_recommendation.csv")
df
```

```python
plt.scatter(df["humidity"],df["temperature"])
plt.xlabel("humidity")
plt.ylabel("temperature")
plt.show()
```

```python
df.rename(columns={"label":"crop"},inplace=True)
```

```python
sns.relplot(x="temperature",y="crop",data=df)
```
