```python
from sklearn import datasets
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
```

**1. Load dataset**

```python
iris = datasets.load_iris()
```

```python
# seperate features and class labels
x_features = iris.data
y_labels = iris.target
```

**2. Model implementation for Hierarchical clsutering**

```python
model = AgglomerativeClustering(linkage = 'ward', n_clusters=3)
```

```python
model.fit(x_features)
predicted_labels = model.labels_
```

```python
predicted_labels
```

**3. Visualizing the Dendrograms**

```python
linkage_matrix = linkage(x_features, 'ward')
plot = plt.figure(figsize=(14,7))
dendrogram(
    linkage_matrix,
    color_threshold=0,
)
plt.title('Hierarchical Clustering Dendrogram (linkage=ward)')
plt.xlabel('sample index')
plt.ylabel('distance')
plt.show()
```

```python
plot_2 = plt.figure(figsize=(14, 7))
dendrogram(
    linkage_matrix,
    truncate_mode='lastp',
    p=20,
    leaf_rotation=90.,
    leaf_font_size=12.,
    show_contracted=True,
)
plt.title('Agglomerative Clustering Dendrogram (linkage=ward)')
plt.xlabel('sample index')
plt.ylabel('distance')
plt.show()
```
